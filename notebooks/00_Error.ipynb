## Error Analysis and Logarithmic Scaling

In numerical methods, it is important not only to compute solutions but also to evaluate their accuracy. This is done by comparing the numerical solution with the exact (analytical) solution.



### Absolute Error

The absolute error is defined as:

$$
\text{Absolute Error} = |x_{\text{num}} - x_{\text{exact}}|
$$

This measures the direct difference between the numerical and exact values.



### Relative Error

The relative error is defined as:

$$
\text{Relative Error} = \frac{|x_{\text{num}} - x_{\text{exact}}|}{|x_{\text{exact}}|}
$$

This provides a scale-independent measure of error and is especially useful when the magnitude of the solution varies significantly.



### Use of Absolute Value

The absolute value is used to ensure that the error represents a magnitude. Without it, positive and negative errors could cancel out, leading to misleading results.
